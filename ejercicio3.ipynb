```python
import pandas as pd
```

```python
df = pd.read_csv("\Users\Usuario\Jhordan\ListaCompanias.csv", index_col="Symbol")
df.head(10)
```

```python
# Transformar los valores string de "MarketCap" a valores numéricos

df["MarketCap"] = df["MarketCap"].str.replace("$","")
df["MarketCap"] = df["MarketCap"].str.replace("M","")
# Transformar los Billones a 10**3(e3) millones
df["MarketCap"] = df["MarketCap"].str.replace("B","e3")
df["MarketCap"] = df["MarketCap"].astype(float)

df.head(10)
```

```python
i = 0
while i <= 4:
    preset = df.sample(n=10)
    set = preset["Sector"].drop_duplicates()
    #print(set.shape)
    if set.shape == (10,):
        print "Portafolio",i+1
        print preset
        print " "
        sumcap = preset['MarketCap'].sum()
        sumcap = float(sumcap)
        print "La suma de las capitalizaciones de mercado es:",sumcap,"millones"
        print " "
        print " "
        i += 1
```
